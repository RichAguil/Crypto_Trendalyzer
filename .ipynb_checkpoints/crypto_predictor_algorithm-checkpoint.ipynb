# Support Vector Regression

The code below will fit and test the support vector regressor, starting off with the RBF kernel

```python
X = bitcoin_train["UTC_timestamp"].values #extracting timestamps
Y = bitcoin_df["price"].values #extracting prices
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2) #splitting data into training and testing set
```

#### RBF Kernel Support Vector Regressor

```python
X_Train = np.reshape(X_Train, (len(X_Train), 1)) #reshaping data so it could be a 1D array
Y_Train = np.reshape(Y_Train, (len(Y_Train), 1))
X_Test = np.reshape(X_Test, (len(X_Test), 1))
SVR_RBF = SVR(kernel= 'rbf', C= 1e5, gamma= 0.001) #creating an instance of the SVR algorithm
SVR_RBF.fit(X_Train, Y_Train) #training the model
```

```python
plt.figure(figsize = (12,6))
plt.plot(X_Train, Y_Train, color= 'black', label= 'Data')
plt.plot(regular_dates, SVR_RBF.predict(X_Test), color= 'red', label= 'RBF model') #predicting and plotting
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

#### Forecasting for the RBF kernalized SVR

```python
forecast = 14 #number of days ahead to forecast
```

```python
prediction = [] # filling list with timedelta values for prediction
for x in range(forecast):
    prediction.append(np.datetime64('2021-01-01') - np.datetime64(-(x+1), 'D'))
print(prediction)
```

```python
prediction = np.reshape(prediction, (len(prediction), 1))
print(svr_rbf.predict(prediction)) #feeding in timedelta values to predict future prices
```

```python
forecast_dates = [] #filling in list with forecasting dates in a regular format
for x in range(forecast):
    forecast_dates.append(np.datetime64('2021-01-01') + np.timedelta64(x, 'D'))
print(forecast_dates)
```

```python
plt.figure(figsize = (12,6))
plt.plot(forecast_dates, SVR_rbf.predict(prediction), color= 'red', label= 'RBF model') 
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
